```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg');
```

Чем больше жилая площадь дома, тем выше его стоимость

```python
counts = df['floors'].value_counts()
    
plt.figure(figsize=(8, 6))    
plt.title('floors')
sns.barplot(x=counts.index, y=counts.values)
    
plt.show()
```

больше всего одноэтажных домов

```python
sns.boxplot(x=df['bedrooms'], whis=1.5, showfliers = False)

plt.xlabel('bedrooms')
plt.show()
```

среднее количество спален 3

Условие 1: Задача 1<br>
Скачать файл kc_house_data.csv - приложен на платформе<br>
Считать данные с помощью pandas<br>
Вывести на экран первые 5 строк<br>
1.2 Создать новый признак delta_renovated, который будет<br> содержать разницу в годах между годом реновацией<br> дома и годом постройки дома<br>
Если реновации дома не было, то в новом признаке поставьте 0<br>
1.3 Создайте признаки года продажи, месяца продажи<br>
1.4 Удалите признаки date, zipcode, lat, long<br>

```python
#1.1
df = pd.read_csv('kc_house_data (2).csv')
df.head()
```

```python
#1.2
new_df = []

for i, k in df.iterrows():
    if k['yr_renovated']:
        new_df.append(k['yr_renovated']-k['yr_built'])
    else:
        new_df.append(0)

df['delta_renovated'] = new_df
df.head()
```

```python
#1.3
df['Y'] = df['date'].apply(lambda x: int(x[:4]))
df['M'] = df['date'].apply(lambda x: int(x[4:6]))
df[['Y','M']].head()
```

```python
#1.4
df.drop(columns=['date' ,'zipcode', 'lat', 'long'], inplace=True)
df.head()
```
